```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
plt.style.use("fivethirtyeight")
```

```python
data=sns.load_dataset("tips")
```

```python
data
```

```python
sns.catplot(x="smoker",y="total_bill",kind="bar",data=data)
```

```python
sns.catplot(x="smoker",y="total_bill",hue="sex",kind="bar",data=data)
```

```python
sns.catplot(x="smoker",y="total_bill",estimator=np.median,kind="bar",data=data)
```

```python
sns.catplot(x="sex",hue="smoker",kind="count",data=data)
```
