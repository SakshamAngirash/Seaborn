```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 

```

```python
plt.style.use("fivethirtyeight")
```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
pt=flights.pivot_table(index="year",columns="month",values="passengers",aggfunc="sum")
```

```python
sns.heatmap(pt,cbar=False)
```

```python
sns.heatmap(pt,cbar=False,linewidth=0.5,annot=True,fmt="d")
```

```python
sns.heatmap(pt,cbar=False,linewidth=0.5,annot=True,fmt="d",cmap="viridis")
```

```python
sns.heatmap(pt,cbar=False,linewidth=0.5,annot=True,fmt="d",cmap="winter")
```

```python
#cluster Mpas
sns.clustermap(pt)
```

```python
sns.clustermap(pt,z_score=0,metric="correlation",row_cluster=False,annot=True)
```
